## Tabla de Drivers

Cargamos las librerias necesarias:

```python
import requests
import json
import pandas as pd
import warnings 
warnings.filterwarnings("ignore")
```

```python
url='http://ergast.com/api/f1/drivers.json?limit=857'
response=requests.get(url)
content=json.loads(response.content)
```

Analizamos la informacion contenida en DriverTable:

```python
content['MRData']['DriverTable']['Drivers']
```

Creamos el diccionario que almacenara la informacion de los conductores:

```python
drivers={'driverId':[],
        'url':[],
        'givenName':[],
        'familyName':[],
        'dateOfBirth':[],
        'nationality':[]}
```

```python
for item in content['MRData']['DriverTable']['Drivers']:
    drivers['driverId'].append(item['driverId'])
    drivers['url'].append(item['url'])
    drivers['givenName'].append(item['givenName'])
    drivers['familyName'].append(item['familyName'])
    drivers['dateOfBirth'].append(item['dateOfBirth'])
    drivers['nationality'].append(item['nationality'])
```

Creamos el dataFrame, o la tabla con el diccionario de drivers:

```python
pd.DataFrame(drivers)
```

Pasamos la tabla de drivers a archivo csv:

```python
pd.DataFrame(drivers).to_csv("drivers.csv", index=False)
```

## Tabla de Circuits

```python
url='http://ergast.com/api/f1/circuits.json?limit=857'
response=requests.get(url)
```

Creamos el diccionario que almacenara la informacion de los circuitos:

```python
circuits={'circuitId':[],
        'url':[],
        'circuitName':[],
        'lat':[],
        'long':[],
        'locality':[],
        'country':[]}
```

```python
pd.DataFrame(circuits).to_csv("circuits.csv", index=False)
```

## Tabla de qualifying

```python
url='http://ergast.com/api/f1/qualifying.json?limit=857'
response=requests.get(url)
```

```python
content=json.loads(response.content)
```

Analizamos la informacion contenida en qualifyng results:

```python
content['MRData']['RaceTable']['Races'][0]['QualifyingResults']
```

Creamos el diccionario que almacenara la informacion de qualifying:

```python
qualifying={'number':[],
        'position':[],
        'driverId':[],
        'url1':[],
        'givenName':[],
        'familyName':[],
        'dateOfBirth':[],
        'nationality1':[],
        'constructorId':[],
        'url2':[],
        'name':[],
        'nationality2':[],
        'Q1':[]}
```

```python
for item in content['MRData']['RaceTable']['Races'][0]['QualifyingResults']:
    qualifying['number'].append(item['number'])
    qualifying['position'].append(item['position'])
    qualifying['driverId'].append(item['Driver']['driverId'])
    qualifying['url1'].append(item['Driver']['url'])
    qualifying['givenName'].append(item['Driver']['givenName'])
    qualifying['familyName'].append(item['Driver']['familyName'])
    qualifying['dateOfBirth'].append(item['Driver']['dateOfBirth'])
    qualifying['nationality1'].append(item['Driver']['nationality'])
    qualifying['constructorId'].append(item['Constructor']['constructorId'])
    qualifying['url2'].append(item['Constructor']['url'])
    qualifying['name'].append(item['Constructor']['name'])
    qualifying['nationality2'].append(item['Constructor']['nationality'])
    qualifying['Q1'].append(item['Q1'])
  
```

Creamos el dataFrame o la tabla de qualifying:

```python
pd.DataFrame(qualifying)
```

Pasamos la tabla de qualifying a archivo csv:

```python
pd.DataFrame(qualifying).to_csv("qualifying.csv", index=False)
```
