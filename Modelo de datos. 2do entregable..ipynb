## Tabla de sprintResults

Cargamos las librerias necesarias:

```python
import requests
import json
import pandas as pd
import warnings 
warnings.filterwarnings("ignore")
```

```python
url='http://ergast.com/api/f1/sprint.json?limit=857'
response=requests.get(url)
content=json.loads(response.content)
```

Analizamos el contenido de SprintResults:

```python
content['MRData']['RaceTable']['Races'][0]['SprintResults']
```

Creamos el diccionaro que almacenara los datos de SprintResults:

```python
sprint={'raceId':[],
        'driverId':[],
        'constructorId':[],
        'number':[],
        'grid':[],
        'position':[],
        'positionText':[],
        'points':[],
        'laps':[],
        'time':[],
        'millis':[],
        'statusId':[],
        'lap':[],
        'Time':[]}
            
```

Añadimos a nuestro diccionario todos los valores de las variables:

```python
for item in content['MRData']['RaceTable']['Races'][0]['SprintResults']:
    sprint["raceId"].append(content['MRData']['RaceTable']['Races'][0]['raceName'])
    sprint['driverId'].append(item["Driver"]["driverId"])
    sprint['constructorId'].append(item["Constructor"]["constructorId"])
    sprint['number'].append(int(item['Driver']['permanentNumber']))
    sprint['grid'].append(int(item["grid"]))
    sprint['position'].append(int(item['position']))
    sprint['positionText'].append(item['positionText'])
    sprint['points'].append(int(item['points']))
    sprint['laps'].append(int(item['laps']))
    sprint['statusId'].append(item['status'])
    sprint['lap'].append(item['FastestLap']["lap"])
    sprint['Time'].append(item['FastestLap']["Time"]["time"])
    try:
        sprin['time'].append(item["Time"]["time"])
    except:
        sprint['time'].append(0)
    try:
        sprint['millis'].append(item["Time"]["millis"])
    except:
        sprint['millis'].append("")
   
    


```

Creamos el dataframe con el diccionario de sprint:

```python
pd.DataFrame(sprint)
```

```python
sprint
```

Pasamos la tabla de sprintResults a archivo csv:

```python
pd.DataFrame(sprint).to_csv("SprintResults.csv", index=False)
```

## Tabla de Status

```python
url='http://ergast.com/api/f1/status.json'
response = requests.get(url)
content=json.loads(response.content)
```

Analizamos la informacion contenida en status:

```python
content['MRData']['StatusTable']
```

Creamos el diccionario que almacenara nuestras variables:

```python
status={'statusId':[],
       'status':[],
       'year':[]}
```

Añadimos la informacion al diccionario, para cada año de 2010 a 2023:

```python
for year in range(2010,2023):
    url='http://ergast.com/api/f1/{}/status.json'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    for item in content['MRData']['StatusTable']['Status']:
        status["statusId"].append(item['statusId'])
        status["status"].append(item['status'])
        status["year"].append(content['MRData']['StatusTable']['season'])
```

Generamos el dataframe del diccionario de status:

```python
pd.DataFrame(status)
```

Pasamos el dataframe a archivo csv:

```python
pd.DataFrame(status).to_csv("Status.csv", index=False)
```

## Tabla Results

```python
url='http://ergast.com/api/f1/results.json?limit=857'
response = requests.get(url)
content=json.loads(response.content)
```

Analizamos el contenido en results:

```python
content['MRData']['RaceTable']['Races'][0]['Results'][0]
```

Creamos el diccionario correspondiente:

```python
results={'raceId':[],
          'driverId':[],
          'constructorId':[],
          'number':[],
          'grid':[],
          'position':[],
          'positionText':[],
          'points':[],
          'laps':[],
          'time':[],
          'millis':[],
          'fastestLap':[],
          'fastestLapSpeed':[],
          'statusId':[],
          'fastestLapTime':[],
          'round':[]}
```

```python
for item in content['MRData']['RaceTable']['Races'][0]['Results']:
    results["raceId"].append(content['MRData']['RaceTable']['Races'][0]['raceName'])
    results["round"].append(int(content['MRData']['RaceTable']['Races'][0]['round']))
    results["driverId"].append(item['Driver']['driverId'])
    results["constructorId"].append(item['Constructor']['constructorId'])
    try:
        results["number"].append(item['Driver']['permanentNumber'])
    except:
        results["number"].append("")
        results["grid"].append(item['grid'])
        results["position"].append(int(item['position']))
        results["positionText"].append(item['positionText'])
        results["points"].append(float(item['points']))
        results["laps"].append(int(item['laps']))
    try:
        results["time"].append(item['Time']['time'])
    except:
        results["time"].append("")
    try:
        results["millis"].append(int(item['Time']['millis']))
    except: 
        results["millis"].append("")
    try:
        results["fastestLap"].append(int(item['FastestLap']['lap']))
    except: 
        results["fastestLap"].append("")
    try:
        results["fastestLapTime"].append(item['FastestLap']['Time']['time'])
    except: 
        results["fastestLapTime"].append("")
    try:
        results["fastestLapSpeed"].append(item['FastestLap']['AverageSpeed']['speed'])
    except: 
        results["fastestLapSpeed"].append("")
                
    try:
        results["statusId"].append(item['status'])
    except: 
        results["statusId"].append("")
                
       
```

Generamos el dataframe de la tabla de results:

```python
pd.DataFrame(results)
```

Pasamos el dataframe a archivo csv:

```python
pd.DataFrame(results).to_csv("Results.csv", index=False)
```
